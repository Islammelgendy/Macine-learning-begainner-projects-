```python
%matplotlib notebook
#importing pandas
import pandas as pd 
#importing apripri and association rule
from mlxtend.frequent_patterns import  apriori
from mlxtend.frequent_patterns import association_rules
```

```python
#reading data
df=pd.read_excel('F:\d.xlsx')
df.head()
```

```python
#getting information about data
df.info()
```

```python
#retun true value if it is null
df.isnull().head()
```

```python
#summing cells with null value
df.isnull().sum()
```

```python
#converting invoice number into string data
df['InvoiceNo'] = df['InvoiceNo'].astype('str')
#creating another data frame with invoiceno that dose not include c
df2 = df[~ df['InvoiceNo'].str.contains('C')]
```

```python
#working on data of the united kingdom only 
basket = df2[df2['Country'] == 'Germany']
```

```python
#sum____sum the total number of sold products
#unstack ___re-ordering data to deal with it
#fillna(0)____replacing null values with "0"
basket = basket.groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack().fillna(0)
```

```python
basket.head()
```

```python
#defining a ne w function callled encode units that return 0 or 1 depend son a condition
def encode_units(x):
    if x <=0:
        return 0
    if x >=1 :
        return 1
```

```python
#applying function on tha data
basket_set = basket.applymap(encode_units)
```

```python
#dropping the cloumn of postage from dat
basket_set.drop('POSTAGE', axis=1, inplace=True)
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
#getting frequent_items with minmum support = .07
frequent_items = apriori(basket_set, min_support= 0.07, use_colnames=True)
frequent_items.head()
```

```python
rules = association_rules(frequent_items, metric='lift', min_threshold=3)
rules
```
